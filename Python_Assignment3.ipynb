# Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()


```python
def myreduce(function,sequence):
    result =sequence[0]
    i =1
    while i< len(sequence):        
        result= function(result,sequence[i])
        i = i +1
    return result  
```

lst = [47,11,42,13]

```python
myreduce(lambda a,b: a + b,lst)
```

```python
def myfilter(function,sequence):
    result =[]
    i =1
    for i in sequence:
        if function(i) == True:
            result.append(i)
        else:
            continue
    return result   
```

```python
lst =[1,2,3,4,5,6,7,8]
```

```python
def odd_check(num):
    if num%2 !=0:
        return True
```

```python
myfilter(odd_check,lst)
```
